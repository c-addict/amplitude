# Амплитудная и частотная модуляция

В радиосвязи высокочастотные олебания используются для переноса информации, такие
колебания называются несущими. Для того чтобы несущее колебание отражало
ередаваемую информацию, необходимо один или несколько араметров несущего
колебания связать с ередаваемым сообщением. Процесс изменения одного или
нескольких параметров несущего колебания по закону передаваемого сообщения
называется модуляция.<br>

В качестве несущегоколебания аиболее часто используют гармонические колебания,
которые описываются формулой<br>

<div style="text-align: center">
    g(t) = U<sub>n</sub>sin(w<sub>n</sub>t+y<sub>n</sub>)
</div>

Параметрами таких сигналов являются амплитуда U<sub>n</sub>, угловая частота
w<sub>n</sub> и начальная фаза y<sub>n</sub><br>

Информационным сигналом s(t) первоначального сообщения можно воздействовать на
любой из параметров несущего колебания. В результате получается три вида модуляции:
амплитудная(АМ), частотная(ЧМ) и фазовая(ФМ). Далее будут рассмотрены только
первые два вида модуляции.<br>

Высокочастотный сигнал, полняющий роль ереносчика информации, называется
переносчиком сообщения, низкочастотный сигнал(информационный сигнал) - модулирующим,
а получаемый ысокочастотный торичный сигнал - модулированным.<br>

## Амплитудная модуляция

При АМ угловая частота w<sub>n</sub> и начальная фаза y<sub>n</sub> несущего
колебания сохраняются неизменными, а с помощью информационного сигнала
s(t) меняется амплитуда несущего колебания U<sub>n</sub>. Амплитуда U(t)<sub>n</sub>
связана  передаваемым сообщением оотношением

<div style="text-align: center">
    U<sub>n</sub>(t) = U<sub>n</sub> + k<sub>A</sub>s(t),
</div>
где U<sub>n</sub> - амплитуда есущего колебания  отсутствии сообщения, k<sub>A</sub> -
коэффициент опорциональности, отражающий степень влияния модулирующего на величину
изменения амплитуды результирующего сигнала, s(t) - функция, зависящая от
времени, соответствующая передаваемому сообщению.<br>

Тогда в аналитической форме АМ-сигнал можно записать так:

<div style="text-align: center">
    g<sub>am</sub>(t) = [U<sub>n</sub>+k<sub>A</sub>s(t)]*sin(w<sub>n</sub>t+y<sub>n</sub>)
</div>

Рассмотрим случай, который называется тональной модуляцией, взяв в качестве
модулирующего сигнала гармоническое олебание.
<div style="text-align: center">
    s(t) = U<sub>m</sub>sin(w<sub>m</sub>t+y<sub>m</sub>)
</div>

```python
import numpy
import matplotlib.pyplot


t = numpy.linspace(0, 1, 100)


try:
    u_m = float(input('Введите амплитуду'))
except ValueError:
    u_m = 1.

try:
    w_m = float(input('Введиде частоту'))
except ValueError:
    w_m = 1.

try:
    y_m = float(input('Введите начальную фазу'))
except ValueError:
    y_m = 1.


s = u_m * numpy.sin(2. * numpy.pi * (w_m + y_m) * t)
```

```python
matplotlib.pyplot.plot(s, 'r')
matplotlib.pyplot.show()
```

Зададим модулируемый сигнал

```python
try:
    u_n = float(input('Введите амплитуду'))
except ValueError:
    u_n = 1.

try:
    w_n = float(input('Введите частоту'))
except ValueError:
    w_n = 1.

try:
    y_n = float(input('Введите начальную фазу'))
except ValueError:
    y_n = 1.


g = u_n * numpy.sin(2 * numpy.pi * (w_n + y_n) * t)
```

```python
matplotlib.pyplot.plot(g, 'g')
```

Тогда АМ-сигнал будет иметь вид:

<div style="text-align: center">
    g<sub>am</sub>(t) = [1 + m<sub>a</sub>sin(w<sub>m</sub>t+y<sub>m</sub>)]U<sub>n</sub>sin(w<sub>n</sub>t+y<sub>n</sub>),
</div>

где m<sub>a</sub> = U<sub>m</sub>/U<sub>n</sub>


```python
m = u_m / u_n

result = (1 + m * numpy.sin(2 * numpy.pi * w_m * t + y_m)) * u_n * numpy.sin(2. * numpy.pi * w_n * t + y_n)
```

```python
matplotlib.pyplot.plot(result)
matplotlib.pyplot.show()
```

## Частотная модуляция

При ЧМ амплитуда несущего колебания U<sub>n</sub> сохраняется постоянной,
а мгновенное значение несущей частоты w(t) определяется модулирующим
сигналом s(t) в соответсвии с выражением:

<div style="text-align: center">
    w(t) = w<sub>n</sub>+k<sub>чм</sub>s(t),
</div>

где k<sub>чм</sub> - коэффициент пропорциональности, связывающий отклонение
изменение w<sub>чм</sub> несущей частоты w(t) от своего номинального значения
w<sub>n</sub>, равное w(t) - w<sub>n</sub>, и величину одулирующего сигнала
s(t), вызывающего то отклонение.

Следовательно, ЧМ-сигнал можно записать так:

<div style="text-align: center">
    g<sub>чм</sub>(t) = U<sub>n</sub>sin[w<sub>n</sub>t + k<sub>чм</sub>s(t)]
</div>


```python
result = u_n * numpy.sin((w_n + ((w_m - w_n) * s)) + y_n)


matplotlib.pyplot.plot(result, 'r')
matplotlib.pyplot.show()
```
