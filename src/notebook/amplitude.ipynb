# Амплитудная модуляция

Амплитудная модуляция - вид модуляции, при которой изменяемым параметром несущего сигнала являтеся его ампплитуда.

Пусть у нас дан сигнал u<sub>c</sub>(t) - несущий(модулируемый) сигнал, вида A<sub>c</sub>*f(w<sub>c</sub>t),
где f = sin или cos

```python
a_c = float(input("Введите амлитуду несущего сигнала"))
w_c = float(input("Введите частоту несущего сигнала"))
f_c = input("Укажите функцию несущего сигнала(sin или cos)")

if f_c != 'sin' or f_c != 'cos':
    f_c = 'cos'
```

u<sub>m</sub>(t) - информационный(модулирующий) сигнал, вида A<sub>m</sub>*f(w<sub>m</sub>t), где f = sin или cos

```python
a_m = float(input("Введите амплитуду информационного сигнала"))
w_m = float(input("Введите частоду информационного сигнала"))
f_m = input("Укажите функцию информационного сигнала(sin или cos)")

if f_m != 'sin' or f_m != 'cos':
    f_m = 'cos'
```

Также дан параметр m - некоторая неотрицательная константа, называемая коэффициентом модуляции.



```python
m = float(input())
```

Тогда результатом применения амплитудной модуляции будет сигнал вида AM(t) = [A<sub>c</sub> + m(t)] * f(w<sub>c</sub>t)

```python
import numpy
import matplotlib.pyplot


t = numpy.linspace(0, 1, 100)


f_c = numpy.sin if f_c == 'sin' else numpy.cos
f_m = numpy.sin if f_m == 'sin' else numpy.cos


carrier = a_c * f_c(2 * numpy.pi * w_c * t)
modulator = a_m * f_m(2 * numpy.pi * w_m * t)
result = a_c * (1 + m * f_c(2 * numpy.pi * w_m * t)) * f_c(2 * numpy.pi * w_c * t)
```

```python
matplotlib.pyplot.plot(modulator, 'g')
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.plot(carrier, 'r')
matplotlib.pyplot.show()
```

```python
matplotlib.pyplot.plot(result, color='purple')
matplotlib.pyplot.show()
```
